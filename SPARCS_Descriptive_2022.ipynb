```python
df = pd.read_csv('https://health.data.ny.gov/resource/5dtw-tffi.csv')
```

```python
selected_columns = ['age_group', 'gender', 'length_of_stay', 'total_charges', 'total_costs', 'type_of_admission']
df_subset = df[selected_columns]
print(df_subset.head())
```

```python
#Make Non-Numeric Numbers Null
df_subset['length_of_stay'] = pd.to_numeric(df_subset['length_of_stay'], errors='coerce')
df_subset['total_charges'] = pd.to_numeric(df_subset['total_charges'], errors='coerce')
df_subset['total_costs'] = pd.to_numeric(df_subset['total_costs'], errors='coerce')
print(df_subset[['length_of_stay', 'total_charges', 'total_costs']].head())

# Mean
mean_length_of_stay = df_subset['length_of_stay'].mean()
mean_total_charges = df_subset['total_charges'].mean()
mean_total_costs = df_subset['total_costs'].mean()
print(f"Mean Length of Stay: {mean_length_of_stay}")
print(f"Mean Total Charges: {mean_total_charges}")
print(f"Mean Total Costs: {mean_total_costs}\n")

# Median
median_length_of_stay = df_subset['length_of_stay'].median()
median_total_charges = df_subset['total_charges'].median()
median_total_costs = df_subset['total_costs'].median()
print(f"Median Length of Stay: {median_length_of_stay}")
print(f"Median Total Charges: {median_total_charges}")
print(f"Median Total Costs: {median_total_costs}\n")

# Standard Deviation
std_length_of_stay = df_subset['length_of_stay'].std()
std_total_charges = df_subset['total_charges'].std()
std_total_costs = df_subset['total_costs'].std()
print(f"Standard Deviation of Length of Stay: {std_length_of_stay}")
print(f"Standard Deviation of Total Charges: {std_total_charges}")
print(f"Standard Deviation of Total Costs: {std_total_costs}\n")

# Min/Max
min_length_of_stay = df_subset['length_of_stay'].min()
max_length_of_stay = df_subset['length_of_stay'].max()
min_total_charges = df_subset['total_charges'].min()
max_total_charges = df_subset['total_charges'].max()
min_total_costs = df_subset['total_costs'].min()
max_total_costs = df_subset['total_costs'].max()
print(f"Min Length of Stay: {min_length_of_stay}, Max Length of Stay: {max_length_of_stay}")
print(f"Min Total Charges: {min_total_charges}, Max Total Charges: {max_total_charges}")
print(f"Min Total Costs: {min_total_costs}, Max Total Costs: {max_total_costs}\n")

#Percentiles & Quartiles
percentiles = [0.25, 0.5, 0.75]
length_of_stay_percentiles = df_subset['length_of_stay'].quantile(percentiles)
total_charges_percentiles = df_subset['total_charges'].quantile(percentiles)
total_costs_percentiles = df_subset['total_costs'].quantile(percentiles)

print(f"Length of Stay Percentiles:\n{length_of_stay_percentiles}\n")
print(f"Total Charges Percentiles:\n{total_charges_percentiles}\n")
print(f"Total Costs Percentiles:\n{total_costs_percentiles}\n")
```

```python
#Distribution of Age Group, Gender, and Type of Admission
age_group_counts = df_subset['age_group'].value_counts()
print("Age Group Distribution:\n", age_group_counts)

# Count the distribution for Gender
gender_counts = df_subset['gender'].value_counts()
print("\nGender Distribution:\n", gender_counts)

# Count the distribution for Type of Admission
type_of_admission_counts = df_subset['type_of_admission'].value_counts()
print("\nType of Admission Distribution:\n", type_of_admission_counts)

import matplotlib.pyplot as plt

#Bar Plot for Age Group
plt.figure(figsize=(10, 5))
plt.bar(age_group_counts.index, age_group_counts.values, color = 'green')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar Plot for Gender
plt.figure(figsize=(6, 4))
plt.bar(gender_counts.index, gender_counts.values, color = 'pink')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Bar Plot for Type of Admission
plt.figure(figsize=(8, 5))
plt.bar(type_of_admission_counts.index, type_of_admission_counts.values, color = 'yellow')
plt.title('Type of Admission Distribution')
plt.xlabel('Type of Admission')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Histogram of Length of Stay
plt.figure(figsize=(10, 5))
plt.hist(df_subset['length_of_stay'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Length of Stay')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()

#Boxplot for Total Charges
plt.figure(figsize=(8, 5))
plt.boxplot(df_subset['total_charges'].dropna(), vert=False, patch_artist=True)
plt.title('Boxplot of Total Charges')
plt.xlabel('Total Charges ($)')
plt.grid(True)
plt.tight_layout()
plt.show()

#Bar plot for Type of Admission to analyze admission trends(Trauma)
plt.figure(figsize=(8, 5))
plt.bar(type_of_admission_counts.index, type_of_admission_counts.values, color='green')
plt.title('Distribution of Type of Admission')
plt.xlabel('Type of Admission')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True, axis='y')
plt.show()
```
